```python
import re
import os
import numpy as np
import pandas as pd
```

```python
# Convert XML files to TXT

input_directory = 'txt/'

for file in os.listdir(input_directory):
    base = os.path.splitext(file)[0]
    os.rename(input_directory + file, input_directory + base + '.txt')
```

```python
def line_matching(directory, row):
    
    line_number = 0
    
    for subdir, dirs, files in os.walk(directory):
        for file in files:
            
            f = open(subdir + '\\' + file, encoding='utf8')
            
            for line in f:
                line_number = line_number + 1
                if "@BCBIdentifierOriginalSourceCode" in line:
                    try:
                        subfolder = re.search("SubFolder = \"(.*?)\"", line).group(1)
                        filename = re.search("FileName = \"(.*?)\"", line).group(1)
                        startline = re.search("StartLine = (.*?),", line).group(1)
                        endline = re.search("EndLine = (.*?)\)", line).group(1)
                        
                        stubberStartline = line_number + 1
        
                    except Exception as e:
                        try:
                            subfolder = re.search("SubFolder=\"(.*?)\"", line).group(1)
                            filename = re.search("FileName=\"(.*?)\"", line).group(1)
                            startline = re.search("StartLine=(.*?),", line).group(1)
                            endline = re.search("EndLine=(.*?)\)", line).group(1)
                            
                            stubberStartline = line_number
                            print(subdir + '\\' + file)
                            print(line_number)
                            print("Stubber startline: ", stubberStartline)
                        except Exception as e:
                            print(subdir + '\\' + file)
                            print(line_number)
                    
                    stubberEndline = (int(endline) - int(startline)) + stubberStartline
        
                    row.append({'subdirectory': subfolder, 'filename': filename, 
                                'true startline': startline, 'true endline': endline,
                                'startline': stubberStartline, 'endline': stubberEndline})
                    
            line_number = 0
```

```python
cols1 = ['subdirectory', 'filename', 'true startline', 'true endline', 'startline', 'endline']
rows1 = []

line_matching('bcb_reduced', rows1)
```
